```python
#to find probability density function between the marks 60 and 70 of secondary school
def get_pdf_probability(dataset,startrange,endrange):
    #importing the necessary libraries for plotting and statistical calculations
    from matplotlib import pyplot
    from scipy.stats import norm
    # visualizing the distribution ...
    # it uses seaborn to create a histogram and kernel density estimation plot of the dataset
    # so this gives the spread of the distribution with blue color and histogram in green 
    import seaborn as sns
    ax = sns.distplot(dataset,kde=True,kde_kws={'color':'blue'},color='Green')
    #It adds vertical lines on the plot or the spread to indicate the start and end of the range of interest
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    # caluculating the mean and standard deviation of the dataset...
    #Mean tells the average value ...standard deviation tells the deviation from the mean
    sample = dataset
    sample_mean =sample.mean()
    sample_std = sample.std()
    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean, sample_std))
    # It defines the normal distribution based on the mean and standard devaiation
    dist = norm(sample_mean, sample_std)
    
    # It generates list of values within the specified ranges(start and end ) and calculates pdf for each values 
    # in the list done using one liner for statement
    #The PDF represents the probability of observing a specific value in a continuous probability distribution
    #For a normal distribution, the PDF is a bell-shaped curve centered around the mean, with the standard deviation 
    #controlling its spread
    
    
    values = [value for value in range(startrange, endrange)]
    probabilities = [dist.pdf(value) for value in values]    
    #adds the probabilities within a specified range to find the total probability mass with in that specified range
    # it is nothing but the indication of data to fall within the specified range
    prob=sum(probabilities)
    #prints the mean, standard deviation and the total probability mass or density withinspecified range
    # for interpretation purpose
    print("The area between range({},{}):{}".format(startrange,endrange,sum(probabilities)))
    return prob
    #this function get_pdf_probability helps to visualize the dataset, calculates the statistical parameters 
    #and provides insights in to the probability of observing data within a specified range in a normal distribution model
```

```python
get_pdf_probability(dataset['ssc_p'],60,70)
```

```python
get_pdf_probability(dataset['ssc_p'],40,40)
```

```python
#the function stdNBgraph converts the dataset in to standard normal distribution
#and visualize using seaborn
def stdNBgraph(dataset):
        # Coverted to standard Normal Distribution
    import seaborn as sns
#calculates the mean and standard deviation
    mean=dataset.mean()
    std=dataset.std()
#creates list of values by iterating over the dataset
    values=[i for i in dataset]
#calculates the z_score by subtracting the mean and dividing the std deviation
#z_score is a measure of how many standard deviations a data point is from the mean
#Converting the data to Z-scores is a common technique to transform data into a standard normal distribution, 
#where the mean is 0 and the standard deviation is 1.
    z_score=[((j-mean)/std) for j in values]
#using a seaborn to visualise the distribution of the z-score
#this generates a histogram and a kernel distribution estimation plot for the z-score
    sns.distplot(z_score,kde=True)
#calculating the mean of the z scores which should be 0 in standard normal distribution
    sum(z_score)/len(z_score)
#this function stdNBgraph takes the dataset, converts in to z-score to make it to a standard normal distribution
# and visualizing the distribution of the z-score using seaborn
#finally it calculates the mean of the z scores to check the value of the mean close to zero 
# as in a standard normal distribution
```

```python
stdNBgraph(dataset['ssc_p'])
```
